# Decision tree

```python
#importated few libers 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing iris dataset from sklearn 
from sklearn.datasets import load_iris
```

```python
# assing the varable to data set
iris = load_iris()
```

```python
iris
```

```python
#from sklearn importing the Dision tress classfier 
from sklearn.tree import DecisionTreeClassifier
```

```python
Tree_cls = DecisionTreeClassifier()
```

```python
Tree_cls.fit(iris.data,iris.target)
```

```python
from sklearn import tree
```

Visualize a Decision Tree

```python
text_representation = tree.export_text(Tree_cls)
print(text_representation)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(Tree_cls,filled=True)
```
