```python
#loading liberies
import pandas as pd
import numpy as np
import seaborn as sns 
import graphviz
from sklearn.tree import DecisionTreeClassifier , export_graphviz
from sklearn import tree
from sklearn.model_selection import train_test_split, GridSearchCV
from  sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix,roc_auc_score,roc_curve
from six import StringIO
from IPython.display import Image
```

```python
#lodating data
data = "https://raw.githubusercontent.com/zygmuntz/wine-quality/master/winequality/winequality-red.csv"
df = pd.read_csv(data,sep=";")
```

```python
df
```

```python
df.describe().T
```

```python
x = df.drop(columns='quality')
y = df['quality']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30,random_state=355)
```

```python
tree_clf = DecisionTreeClassifier()
tree_clf.fit(x_train,y_train)
```

```python
feature_name = list(x.columns)
feature_name
```

```python

target_output = list(y_train.unique())
target_output
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree , export_text
```

```python
plt.figure(figsize=(50,50))
tree =plot_tree(tree_clf,feature_names=feature_name,class_names=str(target_output),filled=True);
plt.savefig("High resoltion.svg",dpi=300)
```

```python
tree_clf.score(x_test,y_test)
```

```python
#crossvaledation
GridSearchCV()
```

```python
#hyperameter 
grid_par = {
    'criterion': ['gini','entropy'],
    'max_depth': range(2,32,1),
    'min_samples_leaf': range(1,10,1),
    'min_samples_split': range(2,10,1),
    'splitter':['best',"random"]
}
```

```python
#input the paremeters for tuning
grid_serch = GridSearchCV(estimator=tree_clf,param_grid=grid_par,cv=5,n_jobs=-1)
```

```python
grid_serch
```

```python
 #fit the train split
    grid_serch.fit(x_train,y_train)
```

```python
#get the best paremeter you got
best_parameter = grid_serch.best_params_
best_parameter
```

```python
#beforre the hyperperameter tunning
grid_serch.best_score_
```

```python
#feeding best parameter from result
Tree_clf = DecisionTreeClassifier(criterion='gini',max_depth=18,min_samples_leaf=6,min_samples_split=4,splitter='random')
Tree_clf.fit(x_train,y_train)
```

```python
#even though the score reduce we have made robust model
Tree_clf.score(x_test,y_test)
```
